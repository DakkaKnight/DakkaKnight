```python
# Importar bibliotecas
import pandas as pd

# Crear listas de valores
name_list = ['Trucha y Sazón', 'La Colina', 'Hotel Silvia', 'Yikart', 'La Esquina', 'DondElio']
payment_type_list = ['Efectivo y tarjeta', 'Efectivo', 'Cualquier metodo', 'Efectivo y Nequi', 'Efectivo', 'Efectivo']
business_type_list = ['Restaurante', 'Restaurante y Tienda', 'Hotel', 'E-Comerce y Artesanias', 'Tienda', 'Bar']
online_store_list = ['Si', 'No', 'Si', 'si', 'No', 'No']

# Usar un diccionario ({key:value}) con el formato {"str":list}
data_dict = {
    "Nombre": name_list,
    "Método Pago": payment_type_list,
    "Tipo Negocio": business_type_list,
    "Tienda en Linea": online_store_list
}

# Convertir el diccionario en un DataFrame
df = pd.DataFrame(data_dict)

# Imprimir el DataFrame
print(df)
```

```python
df
```

```python
# Crear listas para las nuevas columnas
num_employees_list = [12, 5, 30, 8, 4, 10]  # Número de empleados
years_in_operation_list = [5, 10, 15, 3, 8, 12]  # Años en operación

# Agregar nuevas columnas al DataFrame
df['Número de Empleados'] = num_employees_list
df['Años en Operación'] = years_in_operation_list

# Imprimir el DataFrame actualizado
print("\nDataFrame actualizado:")
print(df)
```

```python
df
```

```python
# Crear una nueva columna basada en la relación entre las dos últimas columnas
df['Empleados por Año'] = df['Número de Empleados'] / df['Años en Operación']
df
```
